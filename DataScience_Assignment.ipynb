# Data Splitting and model training

```python
final_Dataset.head()
```

```python
# Separate the dependent and independent variables
X = final_Dataset.drop('Eye movement type', axis=1)  # Independent variables
Y = final_Dataset['Eye movement type']               # Dependent variable
```

```python
#Splitting the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=42)
```

```python
print(X_train.shape,Y_train.shape)
print(X_test.shape,Y_test.shape)
```

```python
# Naive bayes
ga_model = GaussianNB()
ga_model.fit(X_train,Y_train)
ga_pred = ga_model.predict(X_test)
print(classification_report(Y_test, ga_pred))
```

```python
#Logistic Regression
logr_model = LogisticRegression()
logr_model.fit(X_train,Y_train)
logr_prediction = logr_model.predict(X_test)
```

```python
print(classification_report(Y_test, logr_prediction))
```

# Empathy Score

```python
X1_train = final_question1Dataset.drop('Total Score extended',axis = 1)
Y1_train = final_question1Dataset['Total Score extended']
X1_test = final_question2Dataset.drop('Total Score extended',axis = 1)
Y1_test = final_question2Dataset['Total Score extended']
```

```python
#Linear Regression
# Train the model
lr_model = LinearRegression()
lr_model.fit(X1_train, Y1_train)

# Evaluate the model on the testing set
lr_pred = lr_model.predict(X1_test)
lr_mse = mean_squared_error(Y1_test, lr_pred)
print('MSE:', math.sqrt(lr_mse))
lr_r2_score = r2_score(Y1_test, lr_pred)
print("Empathy score", lr_r2_score)
```

```python
#Random Forest
# Train the model
rf_model = RandomForestClassifier()
rf_model.fit(X1_train, Y1_train)
rf_pred = rf_model.predict(X1_test)

lr_mse = mean_squared_error(Y1_test, rf_pred)
print('MSE:', math.sqrt(lr_mse))
rf_r2_score = r2_score(Y1_test, rf_pred)
```

```python
print("Empathy score", rf_r2_score)
```
